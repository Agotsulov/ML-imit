Дана матрица целых чисел А. Полагаем, что число в последней строке последнего столбца "испорчено". Восстановить значение в ячейке таблицы с помощью коэффициента корреляции столбцов (с помощью формулы с лекции)

```python
import numpy as np
import math

def s(P):
    n = np.size(P,0)

    meanP = P.mean()

    #print(P)
    #print(meanP)
    #print(n)

    sumP = 0

    for i in P:
        sumP += (i - meanP) * (i - meanP)    

    #print(sumP)

    return math.sqrt(1/(n-1) * sumP)


def KK(P, Q):
    n = np.size(P,0)

    #print(P)
    #print(Q)
    #print(n)

    meanP = P.mean()
    meanQ = Q.mean()

    #print(meanP)
    #print(meanQ)

    sumPQ = 1

    for i in range(0,n):
        sumPQ += P[i] * Q[i]

    #print(sumPQ)

    sP = s(P) #Неэффективно это каждый раз пересчитывать
    sQ = s(Q)

    #print(sP)
    #print(sQ)

    #print(((sumPQ - n * meanP * meanQ) / ((n - 1) * sP * sQ)))
    return ((sumPQ - n * meanP * meanQ) / ((n - 1) * sP * sQ))


def restore_last_elem(A):
    
    s = np.size(A, 0)
    c = np.size(A, 1)
    
    print(A)

    a = np.delete(A, -1, 0)

    #print(a)

    bad_col = a[:,-1]
    bad_row = A[-1,:]

    print("Плохая строка")
    print(bad_row)

    array_KK = np.zeros(c - 1,float)

    mean_bad_col = bad_col.mean()

    mean_a = np.zeros(c - 1,float)

    for i in range(0,c - 1):
        array_KK[i] = KK(bad_col, a[:,i])
        mean_a[i] = a[:,i].mean()

    print("КК")
    print(array_KK)
    print("Средние кроме последнего столбца")
    print(mean_a)

    up = 0
    down = 0

    for i in range(0,c - 1):
        up += array_KK[i] * (A[-1][i] - mean_a[i])

    for i in array_KK:
        down += math.fabs(i)

    result = mean_bad_col + up / down

    print("Результат")
    print(result)

    A[-1][-1] = result 

    #print(A)
    
    return A


example_from_lecture = np.array([[99,56,91,160,58],[89,58,89,157,48],[91,64,91,165,54],[91,51,91,170,54],[86,56,84,157,44],[97,53,86,175,56],[94,51,91,165,54]])
example_random = np.random.randint(0, 10, size = (10,10))

print("Пример из лекции")
print(restore_last_elem(example_from_lecture))
print("Пример случайного массива")
print(restore_last_elem(example_random))

```

Найти в заданной матрице целых чисел А все строки в которых есть хотябы один 0. Удалить такие строки из матрицы.

```python
import numpy as np #Это каждый раз писать не надо, но я каждую задачу писал и запускал отдельно

s = 10
c = 8

A = np.random.randint(0, 10, size = (s,c))

'''
Наверно все это можно сделать красиво через np.delete() ,но у меня получлось только строки состоящие из нулей удалять
Так что сделал как обычно бы
'''

origin_A = A

print(origin_A)

def index_row_with_zero(a):
    for i in range(0,np.size(a,0)):
        for j in range(0,np.size(a,1)):
            if a[i][j] == 0:
                return i
    return -1
    
index = index_row_with_zero(A)

while index != -1:
    A = np.delete(A, index, 0)
    index = index_row_with_zero(A)

print(A)
```

Дана матрица целых чисел А. Для каждого ее столбца найти среднее арифметическое всех ненулевых элементов. ЗАменить все 0 в матрице на полученное ранее среднее значение (для каждого столбца оно своё).

```python
import numpy as np

s = 10
c = 8

A = np.random.randint(0, 10, size = (s,c))

means = np.zeros(c,float)

print(A)

for i in range(0,c):
    a = A[:,i] #Так достаем стольбцы
    means[i] = a.mean()
    
print(means)

'''Можно в один цикл всё сделать'''
for i in range(0,c):
    for j in range(0,s):
        if A[j][i] == 0:
            A[j][i] = means[i]
            
print(A)
```
