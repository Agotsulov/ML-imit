```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
df = pd.read_csv('girls.csv', sep=';')
df.head()
```

```python
%matplotlib inline
df.plot(y='W', kind='hist', color='red',  title='Weight')
```

```python
df.plot(kind='scatter', y='W', x='L', title='W vs L')
df.plot(kind='scatter', y='W', x='G', title='W vs G')
df.plot(kind='scatter', y='W', x='T', title='W vs T')
df.plot(kind='scatter', y='W', x='B', title='W vs B')
```

```python
df.plot(kind='scatter', y='W', x='G', title='W vs G')
df.plot(kind='scatter', y='W', x='B', title='W vs B')
```

```python
#так создается модель линейной регрессии
#Мы ее запустим пока БЕЗ разделения на тернировочную и тестовую выборки
from sklearn.linear_model import LinearRegression
y=df['W']
X=df.iloc[:,2:6]
lr = LinearRegression()
lr.fit(X, y)
X.head()
```

```python
#так можно посмотреть на ее коэффициенты
lr.coef_
# мы видим, что вс коэфф положительны. Что  это может значить?
```

```python
#что предскажет модель на объектах, которые ей подавались при построении?
y_pred=lr.predict(X)
y_pred
# получаем вектор предсказанных значений
```

```python
# сравнивая истинные значения и предсказанные, можно посчитать MAE, MAPE
print('MAE:',np.sum(abs(y-y_pred))/y.shape[0])
print('MAPE:',100*np.sum(abs(y-y_pred)/abs(y))/y.shape[0])
```

```python
#изобразим ошибку предсказания на каждом объекте
plt.figure(figsize=(16,8))

plt.plot(range(0, y.shape[0]), y)
plt.plot(range(0, y_pred.shape[0]), y_pred)
plt.legend(["true value", "predicted"])
```

Ну а теперь давайте по серьезному: разобьем данные на тестовую и тренир выборки и запустим модель

```python
#выше были определены массивы X, y . Разобем их на две выборки
#в тренировочную выборку пойдут 70% первых объектов
bound=int(y.shape[0]*0.7)
print(bound)
X_train=X[:bound]
X_test=X[bound:]
y_train=y[:bound]
y_test=y[bound:]
```

```python
#запускаем модель на тренировоч выборке
lr = LinearRegression()
lr.fit(X_train, y_train)
#проверяем его качество на тесте
y_pred=lr.predict(X_test)
#сравниваем y_pred y_test, считая для них мае и мапе
print('MAE:',np.sum(abs(y_test-y_pred))/y_test.shape[0])
print('MAPE:',100*np.sum(abs(y_test-y_pred)/abs(y_test))/y_test.shape[0])
```

# Лассо

```python
from sklearn.linear_model import Lasso

y=df['W']
X=df.iloc[:,2:6]

bound=int(y.shape[0]*0.7)
X_train=X[:bound]
X_test=X[bound:]
y_train=y[:bound]
y_test=y[bound:]

```

```python
lasso = Lasso(alpha=1)#константа для лассо здесь называется "альфа" 
lasso.fit(X_train, y_train)
#Тут можно поиграться: увеличивайте значение параметра лассо и смотрите, что будет с коэффициентами
print(lasso.coef_)
y_pred=lasso.predict(X_test)
```

```python
print('MAE:',np.sum(abs(y_test-y_pred))/y_test.shape[0])
print('MAPE:',100*np.sum(abs(y_test-y_pred)/abs(y_test))/y_test.shape[0])
```

Как же выбрать оптимальное значение параметра для лассо?

```python
# создаем массив значений параметра
par_array=np.linspace(0,5,num=50)
print(par_array)
#в цикле строим модель запоминаем ее качество
mae_array = np.zeros(par_array.shape[0])

i = 0
for alpha in par_array:
    lasso = Lasso(alpha = alpha)
    lasso.fit(X_train, y_train)
    y_pred=lasso.predict(X_test)
    mae_array[i]=np.sum(abs(y_test-y_pred))/y_test.shape[0]
    i += 1
print(mae_array)
```

```python
#можно построить график зависимости мае от значения параметра лассо
plt.plot(par_array, mae_array)
plt.xlabel('alpha')
plt.ylabel('mae')
plt.title('Lasso')
```

```python
# найдем лучшее значение параметра лассо
best_index=np.argmin(mae_array)
print('best parameter=',par_array[best_index])
print('mae of best parameter=',mae_array[best_index])
```

# Регуляризация

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1)
ridge.fit(X_train, y_train)
print(ridge.coef_)
y_pred=ridge.predict(X_test)

print('MAE:',np.sum(abs(y_test-y_pred))/y_test.shape[0])
print('MAPE:',100*np.sum(abs(y_test-y_pred)/abs(y_test))/y_test.shape[0])
```

# Как зависят веса признаков от входного парметра лассо и регуляризации?

```python
par_array=np.linspace(0,100,num=50)
coef_lasso = np.zeros((par_array.shape[0], X.shape[1])) #будем запоминать коэфф модели для каждого значения параметра
coef_ridge = np.zeros((par_array.shape[0], X.shape[1]))
i = 0
for alpha in par_array:
    lasso=Lasso(alpha = alpha)
    ridge=Ridge(alpha=alpha)
    lasso.fit(X,y)
    ridge.fit(X,y)
    coef_lasso[i,:]=lasso.coef_
    coef_ridge[i,:]=ridge.coef_
    i += 1
```

```python
plt.figure(figsize=(10, 8))
for coef, feature in zip(coef_lasso.T, df.columns[2:]):
    plt.plot(par_array, coef, label=feature, color=np.random.rand(3), linewidth=3)
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("alpha")
plt.ylabel("feature weight")
plt.title("Lasso")
```

```python
plt.figure(figsize=(10, 8))
for coef, feature in zip(coef_ridge.T, df.columns[2:]):
    plt.plot(par_array, coef, label=feature, color=np.random.rand(3), linewidth=3)
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("alpha")
plt.ylabel("feature weight")
plt.title("Ridge")
```

# Полиномиальная регрессия
Будем искать зависимость веса от других признаков с помощью квадратичной функции. С точки зрения стереометрии - это логично. 
Переход от линейной функции к квадратичной (как и было указано в лекциях) осуществляется с помощью добавления новых признаков

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)# можно уеличивать степень - и посмотреть насколько изменится точность
X_poly=poly.fit_transform(X)
print('new number of columns:',X_poly.shape[1])
print('old number of columns:',X.shape[1])
```

```python
X_train=X_poly[:bound]
X_test=X_poly[bound:]
y_train=y[:bound]
y_test=y[bound:]

#запускаем модель на тренировоч выборке
lr = LinearRegression()
lr.fit(X_train, y_train)
#проверяем его качество на тесте
y_pred=lr.predict(X_test)
#сравниваем y_pred y_test, считая для них мае и мапе
print('MAE:',np.sum(abs(y_test-y_pred))/y_test.shape[0])
print('MAPE:',100*np.sum(abs(y_test-y_pred)/abs(y_test))/y_test.shape[0])
```
