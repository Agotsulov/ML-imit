# Визуализация данных в Питоне
В Python основной библиотекой для рисваония графиков является Matplotlib.
https://matplotlib.org/ можете перейти по ссылке и посмотреть Examples & Tutorials

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
df = pd.read_csv('data_csv.csv', sep=';')
df.head()
```

```python
# Чтобы графики выводились в  прямо jupyter надо выполнить эту магическую команду
%matplotlib inline
```

```python
sns.distplot(df['age'])#так выводятся плотности распределения
```

```python
sns.distplot(df['growth'])#так выводятся плотности распределения
```

```python
#распределение категориальных признаков
sns.countplot(x='sex', data=df)
```

```python
sns.countplot(x='hostel', data=df)
```

```python
sns.countplot(x='coin', data=df)
```

```python
sns.countplot(x='stones', data=df)
```

```python
sns.countplot(y='stones', data=df, hue='sex')
```

```python
sns.countplot(y='sex', data=df, hue='stones')#переставим признаки местами
```

```python
sns.lmplot(x='age', y='growth', hue='sex',data=df)
```

```python
plt.scatter(x=df['EGEmath'],y=df['EGEphis'])#просто рисует пары значений признаков
```

```python
sns.jointplot(x='EGEmath', y='EGEphis', data=df)# все что связано с корреляцией
```

Ящик с усами или диаграмма размаха - компактное отображение распределения вероятности. Показывает медиану, нижний и верхний квантили, минимальное и максимальное значение выборки и выбросы (это те элементы, которые выходят за усы).

```python
sns.boxplot(x='age', data=df)
```
