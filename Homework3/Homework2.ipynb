найти сумму баллов по трем экзаменам ЕГЭ для мужчин и для женщин. Какой пол в среднем имеет лучшую сумму балов? Подумайте над оптимальным порядком суммирования здесь.

```python
print("Способ 1")
m = df[(df['sex'] == 1)]
w = df[(df['sex'] == 0)]
print(m)
print(w)
mean_m = m['EGEmath'].mean() + m['EGEruss'].mean() + m['EGEphis'].mean()
mean_w = w['EGEmath'].mean() + w['EGEruss'].mean() + w['EGEphis'].mean()
print("M")
print(mean_m)
print("W")
print(mean_w)

print("Способ 2")
EGEmath_m = m['EGEmath'].values
EGEruss_m = m['EGEruss'].values
EGEphis_m = m['EGEphis'].values

EGEmath_w = w['EGEmath'].values
EGEruss_w = w['EGEruss'].values
EGEphis_w = w['EGEphis'].values

count_m = np.size(EGEmath_m)
count_w = np.size(EGEmath_w)

sum_m = np.zeros(count_m, int)
sum_w = np.zeros(count_w, int)

for i in range(0, count_m):
    sum_m[i] = EGEmath_m[i] + EGEruss_m[i] + EGEphis_m[i]
    
for i in range(0, count_w):
    sum_w[i] = EGEmath_w[i] + EGEruss_w[i] + EGEphis_w[i]

print("Sum m")
print(sum_m)
print("Sum w")
print(sum_w)
    
print("M")
print(sum_m.mean())
print("W")
print(sum_w.mean())
print("Median M")
print(np.median(sum_m))
print("Median W")
print(np.median(sum_w))
```
