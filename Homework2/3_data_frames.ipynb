```python
#для столбцов можно вычилять статистические характеристики
print(df['growth'].mean(),df['growth'].median(),df['growth'].std())
```

```python
# получение подвыборок по некоторому условию!!!
#найдем всех мужчин, которые живут в общаге
df[(df['sex']==1) & (df['hostel']==1)]
```

```python
#найдем всех мужчин, чей рост ниже роста самой высокой женщины
big_woman=df[df['sex']==0]['growth'].max()
print(big_woman)
df[(df['sex']==1) & (df['growth']<big_woman)]
```

```python
import numpy as np
df.groupby(by='stones').agg([np.mean, np.median])
```

```python
df.groupby('stones').count()['sex']#так через одно место считается число элементов в группе
```

```python
pd.crosstab(df['sex'], df['stones'])#совместное распределение двух величин
```

```python
pd.crosstab(df['sex'], df['coin'])
```

```python
corr_matrix=df.corr()#матрица коэфф корреляции
corr_matrix
```

```python
#матрицу корреляций можно нарисовать
%matplotlib inline
import seaborn as sns
sns.heatmap(corr_matrix)
```
